```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
d = {'g SO2/100 g H2O': [0.02, 0.05,0.1,0.15,0.2,0.3,0.5,0.7,1.0,1.5,2.5,5.0,7.5,10.0], 'Presión parcial de SO2 / mmHg': [0.5,1.2,3.2,5.8,8.5,14.1,26,39,59,92,161,336,517,698]}
df = pd.DataFrame(data=d)
```

```python
df["Presión parcial de SO2 / kPa"]=df['Presión parcial de SO2 / mmHg']*133.3224/1000
df['Masa SO2 (kg)']=df['g SO2/100 g H2O']/1000
df['Masa H20 (kg)']=np.zeros_like(df['Masa SO2 (kg)'])+100/1000
df['Moles de SO2 (mol)']=df['Masa SO2 (kg)']/0.0641
df['Moles de H2O (mol)']=df['Masa H20 (kg)']/0.0180
df['Fracción molar de SO2']=df['Moles de SO2 (mol)']/(df['Moles de SO2 (mol)']+df['Moles de H2O (mol)'])
df
```

```python
df['Flujo total (mol/s)']=0.593132/df['Fracción molar de SO2']
df['Flujo de agua (mol/s)']=df['Flujo total (mol/s)']-0.593132
df
```

```python
plt.figure(figsize=(15,10))
plt.plot(df['Flujo de agua (mol/s)'],df['Presión parcial de SO2 / kPa'])
plt.ylabel(r'Presión parcial de $SO_2$ [kPa]',fontsize=14)
plt.xlabel(r'Flujo de $H_2O$ [mol/s]',fontsize=14)
plt.title(r'Presión parcial de $SO_{2}$ en función del flujo de agua',fontsize=12)
plt.show()
```

```python
df["Concentración de SO2 en aire / mol/m^3"]=df["Presión parcial de SO2 / kPa"]*1000/(8.3144621*(273.15+20))
df['Concentración SO2 en agua [mol/m^3]']=df['Moles de SO2 (mol)']/(df['Flujo de agua (mol/s)']*0.01802/997)
df
```

```python
plt.figure(figsize=(12,5))
plt.plot(df['Concentración SO2 en agua [mol/m^3]'],df['Concentración de SO2 en aire / mol/m^3'])
plt.xlabel(r'Concentración de $SO_{2}$ en agua [mol/m^3]',fontsize=14)
plt.ylabel(r'Concentración de $SO_{2}$ en aire [mol/m^3]',fontsize=14)
plt.title(r'Concentración de $SO_{2}$ en aire como función de su concentración en agua',fontsize=14)
plt.show()
```

```python
vecindario_trabajo=(df[1.5 < df['Presión parcial de SO2 / kPa']])[df['Presión parcial de SO2 / kPa']<30.4]
```

```python
plt.figure(figsize=(12,5))
plt.plot(vecindario_trabajo['Concentración SO2 en agua [mol/m^3]'],vecindario_trabajo['Concentración de SO2 en aire / mol/m^3'])
plt.xlabel(r'Concentración de $SO_{2}$ en agua [mol/m^3]',fontsize=14)
plt.ylabel(r'Concentración de $SO_{2}$ en aire [mol/m^3]',fontsize=14)
plt.title(r'Concentración de $SO_{2}$ en aire como función de su concentración en agua',fontsize=14)
plt.show()
```

```python
res=linregress(vecindario_trabajo['Concentración SO2 en agua [mol/m^3]'].values,vecindario_trabajo['Concentración de SO2 en aire / mol/m^3'].values)
print('El valor de H:',res.slope)
```
